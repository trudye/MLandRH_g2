```python
import pandas as pd

# File paths for your CSVs
file_paths = ['/content/Patient_info.csv', '/content/BASELINE_ABP.csv', '/content/SEPSIS.csv']

# Read CSVs into a list of DataFrames
dfs = [pd.read_csv(file) for file in file_paths]
merged_df = pd.merge(dfs[1], dfs[0], on='admissionid', how='outer')
merged_df = pd.merge(merged_df, dfs[2], on='admissionid', how='outer')


```
